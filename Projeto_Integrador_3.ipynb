## Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Carregamento dos Dados

```python
df = pd.read_csv('/content/grade_horaria.csv')
df.head()
```

## Pré-Processamento

```python
df.info()
```

```python
# 1. Tratamento de Dados Faltantes
# Verificar valores ausentes
df.isnull().sum()
```

```python
# Garantir que colunas como "Semestre" sejam inteiras
df['Semestre'] = df['Semestre'].astype(int)
df['Carga Horária'] = df['Carga Horária'].astype(float)
```

```python
df
```

```python

# 6. Análise Estatística e Agregação
# Estatísticas descritivas para carga horária
estatisticas = df['Carga Horária'].describe()
print(estatisticas)
```

```python

# Frequência por formação
frequencia_formacao = df['Formação'].value_counts()
print(frequencia_formacao)
```

```python
df.info()
```

```python
df
```

```python
# 9. Detecção e Tratamento de Outliers
# Exemplo: Remover outliers na carga horária usando o IQR
q1 = df['Carga Horária'].quantile(0.25)
q3 = df['Carga Horária'].quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr
df = df[(df['Carga Horária'] >= limite_inferior) & (df['Carga Horária'] <= limite_superior)]
```

```python

# Distribuição de carga horária
plt.hist(df['Carga Horária'], bins=10)
plt.title('Distribuição de Carga Horária')
plt.xlabel('Carga Horária')
plt.ylabel('Frequência')
plt.show()
```

```python
# Frequência de formação
df['Formação'].value_counts().plot(kind='bar')
plt.title('Frequência por Formação')
plt.xlabel('Formação')
plt.ylabel('Frequência')
plt.show()
```

## EDA

```python
# Dados dos professores e disciplinas
professores_aulas = df["Professor"].value_counts()
top_5_professores = professores_aulas.head(5)

# Criando o gráfico de barras com barras roxas
plt.figure(figsize=(10, 6))
top_5_professores.sort_values().plot(kind="barh", color='purple')
plt.xlabel("Número de Disciplinas")
plt.ylabel("Professores")
plt.title("Top 5 Professores com Mais Disciplinas Lecionadas")
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

```python
# Identificar o professor que mais deu disciplinas
professor_top = professores_aulas.idxmax()

# Filtrar as disciplinas lecionadas por este professor
disciplinas_professor_top = df[df["Professor"] == professor_top]["Disciplina"].value_counts()

# Criar o gráfico de barras para as disciplinas deste professor com o estilo seaborn-v0_8
plt.style.use('seaborn-v0_8')

plt.figure(figsize=(10, 6))
disciplinas_professor_top.sort_values().plot(kind="barh", color='purple')
plt.xlabel("Número de Vezes Lecionada")
plt.ylabel("Disciplinas")
plt.title(f"Disciplinas Lecionadas por {professor_top}")
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Contar o número de disciplinas por professor separadamente para cada campus
professores_por_campus = df.groupby(["Campus", "Professor"]).size().unstack(fill_value=0)

# Filtrar os top 5 professores que mais deram disciplinas em Asa Sul
top_5_asa_sul = professores_por_campus.loc["Asa Sul"].nlargest(5)

# Criar o gráfico de barras para Asa Sul
plt.figure(figsize=(10, 6))
top_5_asa_sul.sort_values().plot(kind="barh", color='tab:purple')
plt.xlabel("Número de Disciplinas")
plt.ylabel("Professores")
plt.title("Top 5 Professores com Mais Disciplinas Lecionadas (Asa Sul)")
plt.tight_layout()
plt.show()

# Filtrar os top 5 professores que mais deram disciplinas em Taguatinga
top_5_taguatinga = professores_por_campus.loc["Taguatinga"].nlargest(5)

# Criar o gráfico de barras para Taguatinga
plt.figure(figsize=(10, 6))
top_5_taguatinga.sort_values().plot(kind="barh", color='tab:blue')
plt.xlabel("Número de Disciplinas")
plt.ylabel("Professores")
plt.title("Top 5 Professores com Mais Disciplinas Lecionadas (Taguatinga)")
plt.tight_layout()
plt.show()
```

```python
# Identificar o professor que mais deu disciplinas em cada campus
professor_top_asa_sul = professores_por_campus.loc["Asa Sul"].idxmax()
professor_top_taguatinga = professores_por_campus.loc["Taguatinga"].idxmax()

# Filtrar as disciplinas lecionadas por este professor em Asa Sul
disciplinas_professor_top_asa_sul = df[(df["Professor"] == professor_top_asa_sul) & (df["Campus"] == "Asa Sul")]["Disciplina"].value_counts()

# Criar o gráfico de barras para Asa Sul
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10, 6))
disciplinas_professor_top_asa_sul.sort_values().plot(kind="barh", color='tab:purple')
plt.xlabel("Número de Vezes Lecionada")
plt.ylabel("Disciplinas")
plt.title(f"Disciplinas Lecionadas por {professor_top_asa_sul} (Asa Sul)")
plt.tight_layout()
plt.show()

# Filtrar as disciplinas lecionadas por este professor em Taguatinga
disciplinas_professor_top_taguatinga = df[(df["Professor"] == professor_top_taguatinga) & (df["Campus"] == "Taguatinga")]["Disciplina"].value_counts()

# Criar o gráfico de barras para Taguatinga
plt.figure(figsize=(10, 6))
disciplinas_professor_top_taguatinga.sort_values().plot(kind="barh", color='tab:blue')
plt.xlabel("Número de Vezes Lecionada")
plt.ylabel("Disciplinas")
plt.title(f"Disciplinas Lecionadas por {professor_top_taguatinga} (Taguatinga)")
plt.tight_layout()
plt.show()
```

```python
# Agrupar os dados para analisar as disciplinas ministradas por cada professor
preferencia = df.groupby(['Professor', 'Disciplina']).size().reset_index(name='Frequência')

# Encontrar a disciplina que o professor ministra com maior frequência
preferencia_max = preferencia.loc[preferencia.groupby('Professor')['Frequência'].idxmax()]
preferencia_max.rename(columns={'Frequência': 'Frequência_Máxima'}, inplace=True)

# Calcular o desempenho com base na evasão docente
desempenho = df.groupby(['Professor', 'Disciplina'])['EvasãoDocente'].mean().reset_index()
desempenho.rename(columns={'EvasãoDocente': 'Taxa_Evasão'}, inplace=True)

# Combinar os resultados de preferência e desempenho
resultados = pd.merge(preferencia_max, desempenho, on=['Professor', 'Disciplina'])

# Ordenar os resultados para melhor análise
resultados = resultados.sort_values(by=['Professor', 'Taxa_Evasão'], ascending=[True, True])

# Exibir os resultados
print(resultados)
```

```python

# Visualizar em gráficos (opcional)
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de preferências por professor
plt.figure(figsize=(12, 6))
sns.barplot(x='Frequência_Máxima', y='Professor', hue='Disciplina', data=resultados)
plt.title('Disciplinas Preferidas por Professores')
plt.xlabel('Frequência')
plt.ylabel('Professor')
# Posicionar a legenda fora do gráfico
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Disciplina')

plt.show()
```

```python
plt.figure(figsize=(12, 8))  # Aumentar tamanho
sns.barplot(x='Taxa_Evasão', y='Professor', hue='Disciplina', data=resultados)
plt.title('Desempenho por Disciplina (Taxa de Evasão)')
plt.xlabel('Taxa de Evasão')
plt.ylabel('Professor')

# Posicionar a legenda fora do gráfico
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Disciplina')

plt.tight_layout()
plt.show()
```

```python
# Agrupar por professor e contar o número de evasões
evasoes_por_professor = df.groupby('Professor')['EvasãoDocente'].sum().reset_index()

# Ordenar os professores pelo número de evasões em ordem decrescente
evasoes_por_professor = evasoes_por_professor.sort_values(by='EvasãoDocente', ascending=False)

# Renomear a coluna para melhor clareza
evasoes_por_professor.rename(columns={'EvasãoDocente': 'Total_Evasões'}, inplace=True)

# Exibir os resultados
print(evasoes_por_professor)

# Visualizar os dados em um gráfico
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))  # Aumentar o tamanho do gráfico
plt.barh(evasoes_por_professor['Professor'], evasoes_por_professor['Total_Evasões'], color='salmon', height=0.7)  # Ajustar height
plt.title('Total de Evasões por Professor', fontsize=16)
plt.xlabel('Total de Evasões', fontsize=14)
plt.ylabel('Professor', fontsize=14)
plt.gca().invert_yaxis()  # Inverter a ordem dos professores no gráfico
plt.tight_layout()  # Ajustar o layout automaticamente
plt.show()
```

```python
# Supondo que o dataframe esteja carregado como 'df'
# Filtrar as evasões de docentes
evasoes = df[df['EvasãoDocente'] == True]

# Contar as ocorrências por disciplina e pegar o top 5
top_5_evasoes_por_disciplina = evasoes['Disciplina'].value_counts().head(5)

# Ajustar tamanho do gráfico
plt.figure(figsize=(12, 6))  # Largura de 12 e altura de 6

# Criar o gráfico de barras para o top 5
sns.barplot(
    x=top_5_evasoes_por_disciplina.index,
    y=top_5_evasoes_por_disciplina.values,
    palette="coolwarm"
)

# Configurações do gráfico
plt.title("Top 5 Disciplinas com Mais Evasões de Professores", fontsize=16)
plt.xlabel("Disciplina", fontsize=14)
plt.ylabel("Número de Evasões", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.tight_layout()  # Ajusta os elementos para caber na figura

# Mostrar o gráfico
plt.show()
```
